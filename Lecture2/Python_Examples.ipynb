## Exercise 1 — Variables and arithmetic


Python stores values in named variables. Python uses `+`, `-`, `*`, `/`, and `**` for arithmetic.

A rectangular room is 6.0 m long, 4.5 m wide, and 3.0 m high.

Calculate:

1. the floor area;
2. the room volume;
3. the total wall area, excluding the floor and ceiling.

### Your task

````python
length = 6.0
width = 4.5
height = 3.0

# floor_area =
# volume =
# wall_area =

# print(...)
````

### Worked solution

````python
length = 6.0
width = 4.5
height = 3.0

floor_area = length * width
volume = length * width * height
wall_area = 2 * length * height + 2 * width * height

print("Floor area =", floor_area, "m²")
print("Volume =", volume, "m³")
print("Wall area =", wall_area, "m²")
````

## Exercise 5 — Lists and simple statistics


A list stores several values. Useful functions include `len`, `sum`, `min`, and `max`.

For the measurements below, calculate the number of measurements, mean, minimum, maximum, and range.

### Your task

````python
measurements = [12.4, 11.8, 13.1, 12.7, 12.2, 13.5, 11.9]

# number =
# mean =
# minimum =
# maximum =
# data_range =
````

## Exercise 6 — Writing a function


A function packages a calculation so it can be reused.

Write a function called `kinetic_energy` that returns

\[
E_k=\frac{1}{2}mv^2.
\]

Use it for a 1200 kg car travelling at 20 m/s and a 0.145 kg ball travelling at 35 m/s.

### Your task

````python
# def kinetic_energy(...):
#     return ...

# car_energy =
# ball_energy =
````

### Worked solution

````python
def kinetic_energy(mass, velocity):
    return 0.5 * mass * velocity**2

car_energy = kinetic_energy(1200, 20)
ball_energy = kinetic_energy(0.145, 35)

print(f"Car kinetic energy = {car_energy:.1f} J")
print(f"Ball kinetic energy = {ball_energy:.1f} J")
````

## Exercise 7 — Loading the `math` module


Python modules contain useful functions. The standard `math` module includes `pi`, `sin`, `cos`, and `sqrt`.

A simple pendulum has approximate period

\[
T=2\pi\sqrt{\frac{L}{g}}.
\]

Calculate the period for \(L=0.25,\ 0.50,\ 1.00,\ 2.00\) m using \(g=9.81\) m/s².

### Your task

````python
# import math

lengths = [0.25, 0.50, 1.00, 2.00]
g = 9.81

# for length in lengths:
#     period = ...
````

### Worked solution

````python
import math

lengths = [0.25, 0.50, 1.00, 2.00]
g = 9.81

for length in lengths:
    period = 2 * math.pi * math.sqrt(length / g)
    print(f"L = {length:.2f} m, T = {period:.3f} s")
````

Matplotlib is a common plotting module.

Plot

## Exercise 10 — An engineering investigation


For an undamped oscillator,

\[
x(t)=A\cos(\omega t), \quad
v(t)=-A\omega\sin(\omega t), \quad
a(t)=-A\omega^2\cos(\omega t).
\]

For \(A=0.05\) m and \(\omega=4\) rad/s:

1. create a time array from 0 to 10 s;
2. calculate displacement, velocity, and acceleration;
3. create three separate graphs;
4. calculate the maximum speed;
5. compare it with the analytical value \(A\omega\).

### Your task

````python
import numpy as np
import matplotlib.pyplot as plt

A = 0.05
omega = 4.0

# t =
# displacement =
# velocity =
# acceleration =
````

### Worked solution

````python
import numpy as np
import matplotlib.pyplot as plt

A = 0.05
omega = 4.0
t = np.linspace(0, 10, 1000)

displacement = A * np.cos(omega * t)
velocity = -A * omega * np.sin(omega * t)
acceleration = -A * omega**2 * np.cos(omega * t)

plt.figure(figsize=(8, 4))
plt.plot(t, displacement)
plt.xlabel("Time [s]")
plt.ylabel("Displacement [m]")
plt.title("Oscillator displacement")
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 4))
plt.plot(t, velocity)
plt.xlabel("Time [s]")
plt.ylabel("Velocity [m/s]")
plt.title("Oscillator velocity")
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 4))
plt.plot(t, acceleration)
plt.xlabel("Time [s]")
plt.ylabel("Acceleration [m/s²]")
plt.title("Oscillator acceleration")
plt.grid(True)
plt.show()

numerical_max_speed = np.max(np.abs(velocity))
analytical_max_speed = A * omega

print(f"Numerical maximum speed = {numerical_max_speed:.5f} m/s")
print(f"Analytical maximum speed = {analytical_max_speed:.5f} m/s")
print(f"Difference = {abs(numerical_max_speed-analytical_max_speed):.2e} m/s")
````
