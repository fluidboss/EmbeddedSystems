## 1. The central idea

A **class** is a definition or blueprint. An **object** is a particular instance constructed from that blueprint.

A useful analogy is a building plan and the houses built from it. One plan can be used to construct several houses. The houses share a common structure, but each can have different occupants, colours, and addresses.

> **Teaching prompt:** Ask students for another blueprint/instance analogy before showing any code. Possible answers include a questionnaire and completed responses, a recipe and individual cakes, or a CAD design and manufactured components.

## 2. First example: students

The `Student` class below describes the data and behaviour that every student object will have.

Before running the cell, ask the class:

- How many objects will be created?
- Which values will be different between the objects?
- What do you expect `student1.introduce()` to return?

````python
class Student:
    def __init__(self, name, course):
        self.name = name
        self.course = course

    def introduce(self):
        return f"My name is {self.name}. I study {self.course}."


student1 = Student("Alice", "Engineering")
student2 = Student("Ben", "Computer Science")

print(student1.name)
print(student1.introduce())
print(student2.introduce())
````

### Reading the example

- `Student` is the **class**.
- `student1` and `student2` are separate **objects**.
- `name` and `course` are **instance attributes**: each object has its own values.
- `introduce()` is an **instance method**: behaviour provided by the class.
- `__init__()` is called automatically when an object is constructed.
- `self` refers to the particular object currently being initialised or used.

> **Important misconception:** `self` is not a special global variable, and it does not mean the class. It refers to the object on which the method was called.

### What Python effectively does

When we write:

```python
student1.introduce()
```

Python passes `student1` to the method as `self`. The following two calls are therefore equivalent.

````python
print(student1.introduce())
print(Student.introduce(student1))
````

### Objects have independent state

Changing one object does not normally change another object. Predict the output before running the next cell.

````python
student1.course = "Mechanical Engineering"

print(student1.introduce())
print(student2.introduce())
print("Are these the same object?", student1 is student2)
````

## 3. Engineering example: a temperature sensor

This example adds methods that update state, calculate a derived value, and display a formatted result.

> **Teaching prompt:** Ask students to separate the class into *data* and *behaviour*. Data: sensor ID and temperature. Behaviour: update, convert, and display.

````python
class TemperatureSensor:
    def __init__(self, sensor_id, temperature=0.0):
        self.sensor_id = sensor_id
        self.temperature = temperature

    def update_temperature(self, value):
        self.temperature = value

    def temperature_fahrenheit(self):
        return self.temperature * 9 / 5 + 32

    def display(self):
        print(f"Sensor {self.sensor_id}: {self.temperature:.1f} °C")


sensor1 = TemperatureSensor("LAB-01", 21.5)
sensor2 = TemperatureSensor("LAB-02")

sensor1.display()

sensor2.update_temperature(24.2)
sensor2.display()

print(f"Sensor 1: {sensor1.temperature_fahrenheit():.1f} °F")
````

### Points to emphasise

1. `temperature=0.0` is a **default argument**. It is used for `sensor2` because no starting temperature is supplied.
2. `update_temperature()` changes the object's state. It does not need to return the new temperature.
3. `temperature_fahrenheit()` calculates and returns a value without changing the stored Celsius temperature.
4. `display()` handles presentation. In larger programs, calculation and presentation are often kept separate.

> **Discussion question:** Should an impossible temperature be accepted? This opens a useful conversation about validation and protecting object state.

## 4. Class attributes and instance attributes

An **instance attribute** belongs to one object. A **class attribute** is defined on the class and is shared as a default by all its objects.

````python
class BLEDevice:
    connection_type = "Bluetooth Low Energy"

    def __init__(self, name, address):
        self.name = name
        self.address = address

    def describe(self):
        return f"{self.name} at {self.address}"


device1 = BLEDevice("Temperature Sensor", "AA:BB:CC:11:22:33")
device2 = BLEDevice("Heart Rate Monitor", "DD:EE:FF:44:55:66")

print(device1.describe())
print(device2.describe())
print(device1.connection_type)
print(device2.connection_type)
print(BLEDevice.connection_type)
````

### Inspecting where attributes are stored

`vars(object)` shows the object's instance attributes. Notice that `connection_type` is not stored separately in either object. Python finds it on the class.

````python
print("device1 instance attributes:", vars(device1))
print("device2 instance attributes:", vars(device2))
print("Shared class attribute:", BLEDevice.connection_type)
````

## 7. Summary

- A class defines a type of thing.
- An object is a particular instance of that class.
- Instance attributes store each object's state.
- Methods define behaviour and receive the current object as `self`.
- `__init__()` establishes the initial state of a new object.
- Class attributes represent information shared by every instance.
