```python
def dydx(x, y):
    """Defines the ODE dy/dx = x*y."""
    return x * y


def euler_solver(ode_function, x0, y0, x_end, step_size):
    """Solve a first-order ODE using Euler's method."""
    xs = [x0]
    ys = [y0]

    x = x0
    y = y0

    while x < x_end:
        y = y + step_size * ode_function(x, y)
        x = x + step_size

        xs.append(x)
        ys.append(y)

    return xs, ys


# Solve dy/dx = xy, with y = 1 at x = 0
xs, ys = euler_solver(
    ode_function=dydx,
    x0=0.0,
    y0=1.0,
    x_end=1.0,
    step_size=0.01
)

for x, y in zip(xs[::10], ys[::10]):
    print(f"x = {x:.2f}, y = {y:.5f}")
```
