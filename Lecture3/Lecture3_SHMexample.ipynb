This is a code to look at the SHM


```python
import matplotlib.pyplot as plt

# Parameters
omega = 2.0
dt = 0.001
t_end = 10.0

# Initial conditions: x = 0 and y = 1 at t = 0
t = 0.0
x = 0.0
y = 1.0

# Lists for storing results
times = [t]
xs = [x]
ys = [y]

# Euler method
while t < t_end:
    # Calculate both new values from the old values
    x_new = x + dt * y
    y_new = y - dt * omega**2 * x

    t = t + dt
    x = x_new
    y = y_new

    times.append(t)
    xs.append(x)
    ys.append(y)

# Plot x and y against time
plt.plot(times, xs, label="x(t)", color="#2563eb")
plt.plot(times, ys, label="y(t)", color="red")

plt.xlabel("Time, t")
plt.ylabel("x and y")
plt.title("Simple harmonic motion using Euler's method")
plt.grid(True)
plt.legend()
plt.show()
```

Now we plot the data


```python
import math

exact_x = [0.5 * math.sin(2 * t) for t in times]

plt.plot(times, xs, label="Euler solution")
plt.plot(times, exact_x, "--", label="Exact solution")
plt.xlabel("Time, t")
plt.ylabel("Displacement, x")
plt.grid(True)
plt.legend()
plt.show()
```
