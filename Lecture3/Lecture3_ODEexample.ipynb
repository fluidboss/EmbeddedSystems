This is an example of a code used to solve a 1D ODE equation

dy/dx = x*y

Solution is 

y=exp(x^2/2)





```python
def dydx(y, x):
    """The differential equation dy/dx = x*y."""
    return x * y


# Initial conditions
x0 = 0.0
y0 = 1.0

# Final x value and number of steps
x_end = 1.0
number_of_steps = 100

# Calculate the step size
h = (x_end - x0) / number_of_steps

# Store the results
xs = [x0]
ys = [y0]

x = x0
y = y0

# Euler's method
for step in range(number_of_steps):
    y = y + h * dydx(y, x)
    x = x + h

    xs.append(x)
    ys.append(y)

# Display every tenth result
print("       x          y")
print("----------------------")

for index in range(0, len(xs), 10):
    print(f"{xs[index]:8.3f}   {ys[index]:10.6f}")

print(f"\nFinal result: y({xs[-1]:.1f}) = {ys[-1]:.6f}")
```

We next plot out the data

```python

for index in range(0, len(xs), 10):
    print(f"{xs[index]:8.3f}   {ys[index]:10.6f}")

print(f"\nFinal result: y({xs[-1]:.1f}) = {ys[-1]:.6f}")

import matplotlib.pyplot as plt

plt.plot(xs, ys, color="blue", label="Euler solution")

import numpy as np
xpred=np.linspace(0,1.0,10)
ypred=np.exp(xpred*xpred/2)
plt.plot(xpred,ypred,'ro')


plt.xlabel("x")
plt.ylabel("y")
plt.title("Euler solution of dy/dx = xy")

plt.grid(True)
plt.legend()
plt.show()
```
