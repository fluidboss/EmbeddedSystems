We can rewrite the Euler code using classes



```python
class Oscillator:
    def __init__(self, x0, y0, omega):
        self.x = x0
        self.y = y0
        self.omega = omega
        self.time = 0.0

        self.times = [self.time]
        self.positions = [self.x]
        self.velocities = [self.y]

    def euler_step(self, dt):
        x_new = self.x + dt * self.y
        y_new = self.y - dt * self.omega**2 * self.x

        self.time = self.time + dt
        self.x = x_new
        self.y = y_new

        self.times.append(self.time)
        self.positions.append(self.x)
        self.velocities.append(self.y)

    def solve(self, dt, end_time):
        while self.time < end_time:
            self.euler_step(dt)

        return self.times, self.positions, self.velocities
```

```python
oscillator = Oscillator(
    x0=0.0,
    y0=1.0,
    omega=2.0
)

times, positions, velocities = oscillator.solve(
    dt=0.001,
    end_time=10.0
)
```

```python
import matplotlib.pyplot as plt

plt.plot(times, positions, label="Position")
plt.plot(times, velocities, label="Velocity")

plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Simple harmonic motion")
plt.grid(True)
plt.legend()
plt.show()
```
